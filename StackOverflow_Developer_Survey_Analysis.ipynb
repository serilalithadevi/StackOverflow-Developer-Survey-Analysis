- Explore future technology trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use("ggplot")
```

```python
df = pd.read_csv("survey_data_updated 5.csv")

print("Dataset Loaded Successfully!")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.head()
```

```python
df.info()
```

```python
missing = df.isnull().sum().sort_values(ascending=False)

missing.head(20)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
database_list = []

for item in df["DatabaseHaveWorkedWith"].dropna():
    database_list.extend(item.split(";"))

top_db = Counter(database_list).most_common(10)

db_df = pd.DataFrame(top_db, columns=["Database","Count"])

db_df
```

```python
plt.figure(figsize=(10,6))

plt.bar(db_df["Database"], db_df["Count"])

plt.title("Top 10 Databases Used")

plt.xlabel("Database")

plt.ylabel("Number of Developers")

plt.xticks(rotation=45)

plt.show()
```

```python
platform_list=[]

for item in df["PlatformHaveWorkedWith"].dropna():
    platform_list.extend(item.split(";"))

platform_df=pd.DataFrame(
    Counter(platform_list).most_common(10),
    columns=["Platform","Count"]
)

platform_df
```

```python
plt.figure(figsize=(10,6))

plt.barh(platform_df["Platform"],platform_df["Count"])

plt.title("Top 10 Platforms Used")

plt.xlabel("Developers")

plt.gca().invert_yaxis()

plt.show()
```

```python
framework_list=[]

for item in df["WebframeHaveWorkedWith"].dropna():
    framework_list.extend(item.split(";"))

framework_df=pd.DataFrame(
    Counter(framework_list).most_common(10),
    columns=["Framework","Count"]
)

framework_df
```

```python
plt.figure(figsize=(10,6))

plt.barh(framework_df["Framework"],framework_df["Count"])

plt.title("Top 10 Web Frameworks")

plt.xlabel("Developers")

plt.gca().invert_yaxis()

plt.show()
```

```python
from collections import Counter

platforms = df["PlatformWantToWorkWith"].dropna()

platform_list = []

for item in platforms:
    platform_list.extend(item.split(';'))

top_platforms = Counter(platform_list).most_common(10)

platform_df = pd.DataFrame(top_platforms, columns=["Platform","Count"])

platform_df
```

```python
plt.figure(figsize=(10,6))

plt.barh(platform_df["Platform"], platform_df["Count"], color="steelblue")

plt.title("Top 10 Platforms Developers Want to Work With")
plt.xlabel("Developers")

plt.gca().invert_yaxis()

plt.tight_layout()

plt.show()
```

```python
databases = df["DatabaseWantToWorkWith"].dropna()

database_list = []

for item in databases:
    database_list.extend(item.split(';'))

top_databases = Counter(database_list).most_common(10)

database_df = pd.DataFrame(top_databases, columns=["Database","Count"])

database_df
```

```python
plt.figure(figsize=(10,6))

plt.barh(database_df["Database"], database_df["Count"], color="green")

plt.title("Top 10 Databases Developers Want to Work With")
plt.xlabel("Developers")

plt.gca().invert_yaxis()

plt.tight_layout()

plt.show()
```

```python
age_df = df["Age"].value_counts().sort_index()

age_df
```

```python
plt.figure(figsize=(12,6))

age_df.plot(kind="bar", color="orange")

plt.title("Age Distribution of Developers")
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")

plt.tight_layout()

plt.show()
```

```python
employment = df["Employment"].value_counts().head(10)

employment
```

```python
plt.figure(figsize=(10,6))

employment.plot(kind="bar", color="purple")

plt.title("Employment Status")
plt.xlabel("Employment Type")
plt.ylabel("Respondents")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
remote = df["RemoteWork"].value_counts()

remote
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    remote,
    labels=remote.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Remote Work Distribution")

plt.show()
```

```python
country = df["Country"].value_counts().head(10)

country
```

```python
plt.figure(figsize=(10,6))

country.plot(kind="bar", color="teal")

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Respondents")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
salary = df["ConvertedCompYearly"].dropna()

salary.describe()
```

```python
plt.figure(figsize=(10,6))

plt.hist(salary, bins=30, color="red")

plt.title("Salary Distribution")
plt.xlabel("Annual Salary")
plt.ylabel("Developers")

plt.tight_layout()

plt.show()
```
